# Creating a Helper

```python
# Printing Top 5 Rows
def head(n):
  return n.head()

# Printing Last 5 Rows
def tail(n):
  return n.tail()

# Printing Random 10 Rows
def sample(n):
  return n.sample(10)

# Printing Total Number of Rows and Columns
def shape(n):
  return n.shape

# Priniting Dataset Information
def info(n):
  return n.info()

# Priniting Dataset Columns Name
def columns(n):
  return n.columns

# Priniting Dataset Statistical Information
def describe(n):
  return n.describe()
```

```python
data['total'] = num.sum(axis=1)
```
